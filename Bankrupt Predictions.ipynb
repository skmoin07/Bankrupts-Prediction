```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Correlation
corr=data.corr()
plt.figure(figsize=(25,15))
sns.heatmap(corr,cmap='Blues',annot=None)
```

```python
x=data.drop('Bankrupt?',axis=1)
y=data['Bankrupt?']
```

```python
names=x.columns
```

```python

from sklearn import preprocessing
X=preprocessing.normalize(x, axis=0)
x= pd.DataFrame(X, columns=names)
x.head()
```

```python
x.shape
```

```python
y.shape
```

```python
#Feature Importance
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(x,y)
print(model.feature_importances_)
feature_imp=pd.Series(model.feature_importances_,index=x.columns)
plt.figure(figsize=(30,15))
feature_imp.nlargest(80).plot(kind='bar')
```

```python
feature_imp.nsmallest(12)
```

```python
x.columns
```

```python
plt.figure(figsize=(20,10))
feature_imp.nsmallest(15).plot(kind='bar')
```

```python
x=x.drop([' Current Ratio',' Interest-bearing debt interest rate',' Net Income Flag',' Revenue Per Share (Yuan ¥)',' Quick Assets/Current Liability',' Average Collection Days',' Accounts Receivable Turnover',' Total assets to GNP price',' Allocation rate per person',' Net Value Growth Rate',' Revenue per person',' Quick Ratio'],axis=1)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=22)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Model Building
log=pd.DataFrame(columns=["Model", "Accuracy"])
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix               

model = RandomForestClassifier(n_estimators =50, random_state = 22)
model.fit(x_train, y_train)
y_pred_rf = model.predict(x_test)

print("Training Accuracy: ", model.score(x_train, y_train))
print('Testing Accuarcy: ', model.score(x_test, y_test))

# making a classification report
cr = classification_report(y_test,  y_pred_rf)
print(cr)

# making a confusion matrix
plt.rcParams['figure.figsize'] = (5, 5)
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot = True, cmap = 'Blues')
plt.show()
```

```python
log =log.append({"Model": " Random Forest", "Accuracy":  model.score(x_test, y_test)}, ignore_index=True)
```

```python
#Gradient Boosting 
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

model2 =GradientBoostingClassifier(learning_rate=0.01,random_state=22)
model2.fit(x_train, y_train)
y_pred_rf = model2.predict(x_test)
print("Training Accuracy: ", model2.score(x_train, y_train))
print('Testing Accuarcy: ', model2.score(x_test, y_test))

# making a classification report
cr = classification_report(y_test,  y_pred_rf)
print(cr)

# making a confusion matrix
plt.rcParams['figure.figsize'] = (5, 5)
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot = True, cmap = 'copper')
plt.show()
```

```python
log =log.append({"Model": "Gradient Boosting ", "Accuracy":  model2.score(x_test, y_test)}, ignore_index=True)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

model3 =KNeighborsClassifier(n_neighbors=20)
model3.fit(x_train, y_train)
y_pred_rf = model3.predict(x_test)
print("Training Accuracy: ", model3.score(x_train, y_train))
print('Testing Accuarcy: ', model3.score(x_test, y_test))

# making a classification report
cr = classification_report(y_test,  y_pred_rf)
print(cr)

# making a confusion matrix
plt.rcParams['figure.figsize'] = (5, 5)
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot = True, cmap = 'gist_earth')
plt.show()
```

```python
log =log.append({"Model": "KNN", "Accuracy": model3.score(x_test, y_test)}, ignore_index=True)
```

```python
#DECISION TREE
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
model4 =DecisionTreeClassifier(criterion='entropy',max_depth=5)
model4.fit(x_train, y_train)
y_pred_rf = model4.predict(x_test)
print("Training Accuracy: ", model4.score(x_train, y_train))
print('Testing Accuarcy: ', model4.score(x_test, y_test))

# making a classification report
cr = classification_report(y_test,  y_pred_rf)
print(cr)

# making a confusion matrix
plt.rcParams['figure.figsize'] = (5, 5)
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot = True, cmap = 'viridis')
plt.show()
```

```python
log =log.append({"Model": "DECISION TREE", "Accuracy":  model4.score(x_test, y_test)}, ignore_index=True)
```

```python
#XGBM
import xgboost as xgb
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

model5 =xgb.XGBClassifier(random_state=22,learning_rate=0.01)
model5.fit(x_train, y_train)
y_pred_rf = model5.predict(x_test)
print("Training Accuracy: ", model5.score(x_train, y_train))
print('Testing Accuarcy: ', model5.score(x_test, y_test))

# making a classification report
cr = classification_report(y_test,  y_pred_rf)
print(cr)

# making a confusion matrix
plt.rcParams['figure.figsize'] = (5, 5)
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot = True, cmap = 'bone_r')
plt.show()
```

```python
log =log.append({"Model": "XGBM", "Accuracy":  model5.score(x_test, y_test)}, ignore_index=True)
```
